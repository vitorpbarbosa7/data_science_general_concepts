## Logistic map

```r
logisticmap <- function(r,x){
    return(r*x*(1-x))
    }
```

```r
r = 3.95
init = 0.4
y = c(init)
x = c(0)
years = 1000

for(i in 0:years){
    init = logisticmap(r,init)
    x[i] = i + 1
    y[i] = init
    }
```

```r
fg(15,10)
plot(x,y, pch = 20, col = 2)
lines(x,y)
```

```r
chaos = y
```

```r
mutual(chaos)
```

```r
plot(false.nearest(chaos, m = 10, d = 5, t = 1, eps = 0.5))
```

```r
phase =  embedd(chaos, m = 3, d = 5)
```

# Recurrence Plot

## rp

```r
time = seq(0,9, len =1000)
```

```r
X = sin(2*pi*time)
```

```r
rp <- function(X, m, d, eps){
    # m: numero de eixos do espaco fase 
    # d: time delay
    # X: serie temporal
    
    phase = embedd(X, m=m, d=d)
    R = matrix(0, nrow = nrow(phase), ncol = nrow(phase))
    
    # comparar distancia de cada estado com o outr
    # distancia entre cada linha da matriz de embedding
    for (i in 1:nrow(phase)){
        for (j in 1:nrow(phase)){
            Eij = sqrt(sum((phase[i,] - phase[j,])^2))
            
            # preencher a matrix
            if (Eij < eps){
                # sao epsilon proximos
                R[i,j] = 1
            }
        }
    }
    return(R)
}
```

```r
for (i in c(0.15,0.3,0.5,0.8,1,1.5,3)){
        rmatrix = rp(X = X, m =3,d = 6,eps = 1.5)
        graphics::filled.contour(rmatrix)
    }
```
